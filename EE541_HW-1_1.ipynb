Counting the number of heads in 50 trials

```python
import random # Importing the in-built python module. 

head = 0 
i = 0 # While loop iteration variable.

while i < 50: 
  coin_flip = random.random() # Generating random numbers between 0 and 1.
  if coin_flip >= 0.3: # P[head] = 0.7
    head += 1 # Incrementing the head counter. 
  i += 1
  
print("The number of times the flip resulted in a head = ", head)
```

Recording the longest run of heads in 50 flips

```python
head_list = [] # List of heads and tails resulting from the coin toss.
decision = None # Variable which decides the coin toss result.
i1 = 0 # While loop iteration varaiable.
longest_head = 0 # Variable that stores the longest run of heads.
headlist_count = 0 # Counter within the head_list.
ind_headlist = [] # List containing the number of heads.

while i1 < 50: 
  headcoin_flip = random.random()
  if headcoin_flip >= 0.3:
    decision = 1 # 1 -> head
  else:
    decision = 0 # 0 -> tail
  head_list.append(decision)
  i1 += 1

for ele in head_list:
  if ele == 1:
    headlist_count += 1
  else:   
    ind_headlist.append(headlist_count)
    headlist_count = 0 # Resetting the value after every iteration.
  ind_headlist.append(headlist_count) # Appending the values in the list (outside the else condition) to avoid edge cases.

ind_headlist = [element for element in ind_headlist if element != 0] # Removing the 0's (tails) from the list.
longest_head = max(ind_headlist)

#print(head_list)
#print(ind_headlist)
print("The longest run of heads- ", longest_head)
```

Defining the 50 trial coin flip function

```python
def coinflip_50():
  head_50 = 0
  i50 = 0

  while i50 < 50:
    head50coin_flip = random.random()
    if head50coin_flip >= 0.3:
      head_50 += 1
    i50 += 1
  return head_50
```

Generation of histogram for 20 iteration of 50-flip experiment

```python
import matplotlib.pyplot as plt # Importing the in-built python module for plotting the histogram graph.

headlist_20 = []
for x in range(20):
  headlist_20.append(coinflip_50()) # Calling the "50 trial coin flip" function 20 times.
#print(headlist_20) 
plt.title("Histogram for the 20 iteration of 50-flip experiment \n", fontweight = 'bold')
plt.xlabel('Number of heads recorded in a particular realization')
plt.ylabel('Frequency of heads in the particular realization')
plt.hist(headlist_20)
plt.show()
```

This histogram closely resembles a normal distribution curve. As the coin toss repetition increases the respective histograms come closer to resembling a normal distribution curve (i.e., the curve smoothens out). Also, as the coin toss repetitions tend towards infinity, a proper bell-curved histogram may be achieved. In each of the above histograms, the number of times head has shown up after a coin toss is maximum at the 35th mark of the histogram. This is valid because 70% (P[head]) of 50 (realizations) is 35.
