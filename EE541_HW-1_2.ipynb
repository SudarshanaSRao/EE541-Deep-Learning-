2. Define the random variable N = min {n : Pn i=1 Xi > 4} as the smallest number of standard uniform random samples whose sum exceeds four. Generate a histogram using 100, 1000, and 10000 realizations of N. Comment on the expected value E[N].

```python
# Importing the in-built python module. 
import random
import matplotlib.pyplot as plt

# Generation of histogram for 100 realizations of N
i = 0 # While loop iteration variable.
N = []
summ = 0 # Computing the sum of the randomly generated numbers. 
count = 0 # Counting the minimum numbers when the sum exceeds 4. 

while i < 100:
  while summ <= 4:
    n = random.random() # Generating random numbers between 0 and 1.
    summ += n
    count += 1
  N.append(count)
  # Resetting the values after every iteration.
  summ = 0
  count = 0
  i += 1

avg = sum(N) / len(N)
print("E[n]= ", avg) # Expected value E[n] function.
#print(N)
#print(count)
plt.title("Histogram for 100 realizations of N \n", fontweight = 'bold')
plt.hist(N, color = 'brown')
plt.show()
```

The expected value or E[n] is the mean of N and E[n] lies close to the middle of the histogram. Here the histogram resembles a bell curve or Gaussian/normal distribution.
